```python
#Linegraph dataset cleaning
csvpath2 = os.path.join('Resources','WHR2024.csv')
dataframe24 = pd.read_csv(csvpath2)
dataframe24.head()
```

```python
dataframe_2024_subset= dataframe24[['Country name', 'Ladder score']]
dataframe_2024_subset.head()
```

```python
dataframe_2023 = dataframe[['Country name', 'Ladder score',]]
dataframe_2023.head()
```

```python
#merge 2023 and 2024
# Merge the datasets on the "Country name" column using an outer join and add suffixes for each DataFrame
merged_2023_2024dataframe = dataframe_2024_subset.merge(dataframe_2023, on="Country name", how="outer", suffixes=('_2024', '_2023'))

# Display the merged DataFrame
merged_2023_2024dataframe

```

```python
#get 2022 data
csvpath3 = os.path.join('Resources','WHR2022.csv')
dataframe_2022 = pd.read_csv(csvpath3)

    # Select specific columns
dataframe_2022
```

```python
dataframe_2022_clean = dataframe_2022[['Country','Happiness score',]]
dataframe_2022_clean

#renaming columns
dataframe_2022_clean = dataframe_2022_clean.rename(columns={"Country": "Country name", "Happiness score":"Ladder score_2022"})
dataframe_2022_clean
```

```python
#merge 2022+ 2023 and 2024
# Merge the datasets on the "Country name" column using an outer join and add suffixes for each DataFrame
merged_2022_2024dataframe = merged_2023_2024dataframe.merge(dataframe_2022_clean, on="Country name", how="outer")
merged_2022_2024dataframe


```

```python
#make dataframe of 2021 
csvpath4 = os.path.join('Resources','WHR2021.csv')
dataframe21 = pd.read_csv(csvpath4)
dataframe21.head()

dataframe21_clean = dataframe21[['Country name','Ladder score',]]
dataframe21_clean
```

```python
#renaming 2021 columns

dataframe21_clean = dataframe21_clean.rename(columns={"Ladder score": "Ladder score_2021"})
dataframe21_clean
```

```python
#make dataframe of 2020
csvpath5 = os.path.join('Resources','WHR2020.csv')
dataframe20 = pd.read_csv(csvpath5)
dataframe20.head()


# Select specific columns ('Country name' and 'Ladder score') for the cleaned DataFrame
dataframe20_clean = dataframe20[['Country name', 'Ladder score']]

# Rename the "Ladder score" column to "Ladder score_2020" in dataframe20_clean
dataframe20_clean = dataframe20_clean.rename(columns={'Ladder score':'Ladder score_2020'})
dataframe20_clean
```

```python
#merge 2021+ 2020
# Merge the datasets on the "Country name" column using an outer join and add suffixes for each DataFrame
merged_2020_2021 = dataframe21_clean.merge(dataframe20_clean, on="Country name", how="outer")
merged_2020_2021
```

```python
#merge20 and 21 DF with 2022_2024
merged_2020_2024dataframe = merged_2022_2024dataframe.merge(merged_2020_2021, on="Country name", how="outer")
merged_2020_2024dataframe
```

```python
#make dataframe of 2020
csvpath6 = os.path.join('Resources','WHR2019.csv')
dataframe19 = pd.read_csv(csvpath5)
dataframe19.head()


# Select specific columns ('Country name' and 'Ladder score') for the cleaned DataFrame
dataframe19_clean = dataframe19[['Country name', 'Ladder score']]

# Rename the "Ladder score" column to "Ladder score_2020" in dataframe20_clean
dataframe19_clean = dataframe19_clean.rename(columns={'Ladder score':'Ladder score_2019'})
dataframe19_clean
```

```python
# Load 2018 data
csvpath6 = os.path.join('Resources', 'WHR2018.csv')
dataframe18 = pd.read_csv(csvpath6)



# Clean 2018 data
dataframe18_clean = dataframe18[['Country or region', 'Score']]
dataframe18_clean = dataframe18_clean.rename(columns={
    'Country or region': 'Country name',
    'Score': 'Ladder score_2018'
})

dataframe18_clean
```

```python
merged_2018_2019dataframe = dataframe19_clean.merge(dataframe18_clean, on="Country name", how="outer")
merged_2018_2019dataframe
```

```python
complete_df= merged_2020_2024dataframe.merge(merged_2018_2019dataframe, on="Country name", how="outer")
complete_df.set_index(['Country name'], inplace=True)

complete_df
```

```python
#reorder the dataframe and remove the Ladder_score word
reordered_df = complete_df[['Ladder score_2018', 'Ladder score_2019', 'Ladder score_2020',
                                     'Ladder score_2021', 'Ladder score_2022', 'Ladder score_2023', 
                                     'Ladder score_2024']]
reordered_df.rename(columns=lambda x: x.split('_')[-1], inplace=True)
reordered_df 
```

```python
#clear NA values for 2018, 2019 and 2024 as being used for analysis
reordered_complete= reordered_df.dropna(subset=['2018', '2020', '2024'])
reordered_complete
```

```python
#import as csv

reordered_complete.to_csv("Resources/allcountries_df.csv")
```

```python
#Does Australia, Spain, China and the USA get happier over 2021-2024? clean DF for these countries
selected_countries_df = reordered_complete.loc[['Australia', 'Spain', 'China', 'United States']]
selected_countries_df
```

```python
#export to csv 
selected_countries_df.to_csv("Resources/selected_countries_df.csv")
```
